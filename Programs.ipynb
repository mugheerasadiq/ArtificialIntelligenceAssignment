## Genetic Algorithm code

```python
def genitic_algorithm(number_of_bits, population_size, max_iterations = 1000):

    # initialization
    inititial_population = []
    for i in range(0, population_size):
        genome = []
        for j in range(0, number_of_bits):
            genome.append(random.randrange(2))
        inititial_population.append(genome)

    mutation_population = inititial_population
    iterations = 0
    min_population = []
    min_population_index = 0
    min_val = 2**len(inititial_population[0])
    max_population = []
    max_population_index = 0
    max_val = -1
    while max_val != 2**len(inititial_population[0]) - 1 and iterations < max_iterations:
        iterations += 1
        inititial_population = mutation_population
#         print('size of initial pop' + str(len(inititial_population[0])))
#         print((inititial_population[0]))

        # selection
        min_population = []
        min_population_index = 0
        min_val = 2**len(inititial_population[0])
        max_population = []
        max_population_index = 0
        max_val = -1
        for index,genome in enumerate(inititial_population):
            genome_fitness = fitness_function_convert(deepcopy(genome))
            if genome_fitness < min_val:
                min_val = genome_fitness
                min_population_index = index    
                min_population = genome
            if genome_fitness > max_val:
                max_val = genome_fitness
                max_population_index = index    
                max_population = genome

        # selected_population = sorted(inititial_population, key = lambda x: Counter(x)[1], reverse=True)[:selection_size]
        inititial_population.pop(min_population_index)
        selected_population = inititial_population
        selected_population += [max_population]

        # crossover
        crossover_location = int(len(selected_population[0]) / 2)
        crossover_population = []
        for i in range(0, len(selected_population)):
            genome = []
            if i == 0:
                genome = selected_population[i][:crossover_location]
                genome += selected_population[i+1][crossover_location:]
            else:
                genome = selected_population[i][:crossover_location]
                genome += selected_population[i-1][:crossover_location]
            crossover_population.append(genome)
    #     crossover_population

        # mutation
        mutation_population = crossover_population

        for i in range(0, len(mutation_population)):
        #     print(mutation_population[i])
            mutation_index = random.randrange(len(mutation_population[i])-2)

            mutation_population[i][mutation_index] = random.randrange(2)
        #     print(mutation_population[i])
    #     mutation_population

    return {
        'generations':iterations,
        'max_val':max_val,
        'max_population':max_population,
        'number_of_bits':number_of_bits,
        'population_size':population_size
    }
```

## N Queens - Problem 2

## Helper Functions

```python
def display_board(board):
    for x in board:
        for y in x:
            print(y, end=' ')
        print()

def board_rows(board):
    return len(board)

def board_cols(board):
    return len(board[0])

def place_queen(board, row, col):
    board[row][col] = 1
    
def remove_queen(board, row, col):
    board[row][col] = 0
    
def row_is_clear(board, row, col):
    r = deepcopy(row)
    c = 0
    while c < board_cols(board):
        if(board[row][c] == 1):
            return False
        c += 1
    return True
    

def upper_diag_is_clear(board, row, col):
    r = deepcopy(row)
    c = deepcopy(col)
    while c >=0 and r < board_rows(board):
        if board[r][c] == 1:
            return False
        r += 1
        c -= 1
    return True
def lower_diag_is_clear(board, row, col):
    r = deepcopy(row)
    c = deepcopy(col)
    while r >=0 and col >= 0:
        if board[r][c] == 1:
            return False
        r -= 1
        c -= 1
    return True
    
def is_safe(board, row, col):
    return lower_diag_is_clear(board, row, col) and row_is_clear(board, row, col) and upper_diag_is_clear(board, row, col)
```

## Solving N-Queens Algorithm

```python
def solve_queens(board, col):
    if col >= len(board[0]):
        return True
    else:
        for row in range(0, board_rows(board)):
            clear_output(wait=True)
            if(is_safe(board, row, col)):
                place_queen(board, row, col) # Try Queen at position row, col
                display_board(board)
                time.sleep(0.5)
                if(solve_queens(board, col + 1)): # Check next col
                    return True
                remove_queen(board, row, col) # Failed, Remove queen at row, col and continue loop
        return False
```

## MRV & MCV - Problem 3

Color Class that will accept an Enumeration.

```python
domain = 3
minimumCount = 0

class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3
```

The Graph Class that will make a graph and colors them according to their appropriate domain.

```python
class Graph:
    def __init__(self, totalNodes, adjacencyMatrix, color):
        self.totalNodes = totalNodes
        self.adjacencyMatrix = adjacencyMatrix
        self.color = color
        self.nodeSequence = [""]*totalNodes

    def NodePicker(self, initialNode):
        maxCount = 0
        selectedNode = ''
        if (initialNode == ''):
            for node, neighbourList in self.adjacencyMatrix.items():
                if (len(neighbourList) > maxCount and self.color[node] == 0):
                    maxCount = len(neighbourList)
                    selectedNode = node
        else:
            for i in range(len(self.adjacencyMatrix[initialNode])):
                childNode = self.adjacencyMatrix[initialNode][i]
                if (self.color[childNode] == 0 and len(self.adjacencyMatrix[childNode]) > maxCount):
                    maxCount = len(self.adjacencyMatrix[childNode])
                    selectedNode = childNode

        return selectedNode

    def getNodeWithMRV(self, parentNode, domain):
        selectedNode = ''

        for i in range(len(self.adjacencyMatrix[parentNode])):
            childNode = self.adjacencyMatrix[parentNode][i]
            countColor = 0
            for c in range(1, domain+1):
                if(self.isEligible(childNode, c) == True):
                    countColor += 1
            if (countColor < minimumCount):
                selectedNode = childNode

        return selectedNode

    def isEligible(self, node, c):
        for i in range(len(self.adjacencyMatrix[node])):
            if(self.color[self.adjacencyMatrix[node][i]] == c):
                return False
        return True

    def graphColorHelper(self, node, domain):
        if node == '':
            for key, value in self.color.items():
                if value == 0:
                    self.graphColorHelper(key, domain)
            return True

        for c in range(1, domain+1):
            if(self.isEligible(node, c) == True):
                self.color[node] = c
                nextNode = self.getNodeWithMRV(node, domain)
                if(self.graphColorHelper(nextNode, domain) == True):
                    return True
                else:
                    self.color[node] = 0

        return False

    def graphColoring(self, domain):
        startNode = self.NodePicker('')
        if(self.graphColorHelper(startNode, domain) == True):
            return True
        else:
            print("No solution exists.")
            return False
```
